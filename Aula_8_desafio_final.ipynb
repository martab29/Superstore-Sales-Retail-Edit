```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.dtypes
```

```python
#categoria vs vendas
#quantos dias de shipping demora
#onde temos lucro/prejzo
#quais sao as melhores regioes
```

```python
#categoria vs vendas
category_sales = df.groupby('Category')['Sales'].sum().reset_index().sort_values('Sales', ascending=False)
category_sales
```

```python
df['lead_time'] = (df['Ship Date']-df['Order Date']).dt.days
```

```python
#onde temos lucro/prejzo

category_profit = df.groupby('Category')['Profit'].sum().reset_index().sort_values('Profit', ascending=False)
category_profit
```

```python
#onde temos lucro/prejzo

profit_region = df.groupby('Region')['Profit'].mean().reset_index().sort_values('Profit', ascending=False)
profit_region
```

```python
df.groupby('Region')['lead_time'].mean().reset_index().sort_values('lead_time', ascending=False)
```

# 5. VisualizaÃ§Ã£o

**5.1) Qual a categoria com mais vendas?**

```python
category_sales.plot(x='Category', y='Sales', kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('')
```

```python
category_profit.plot(x='Category', y='Profit', kind='bar')
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
```

```python
category_comparison = pd.merge(category_sales, category_profit, on='Category', suffixes=('_Sales', '_Profit'))
category_comparison_melted = category_comparison.melt(id_vars='Category', var_name='Metric', value_name='Amount')

plt.figure(figsize=(12, 7))
sns.barplot(data=category_comparison_melted, x='Category', y='Amount', hue='Metric', palette={'Sales': 'skyblue', 'Profit': 'lightcoral'})
plt.title('Sales and Profit by Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**5.2) Onde temos mais lucro/prejuÃ­zo**

```python
#Lucro por regiÃ£o

profit_region.plot(x='Region', y='Profit', kind='bar')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')
```

```python
#Lucro por state

profit_state = df.groupby("State")["Profit"].sum().sort_values()

plt.figure(figsize=(15,10))
sns.barplot(x=profit_state.values, y=profit_state.index)
plt.title("Profit by State")
plt.xlabel("Profit")
plt.ylabel("State")
plt.show()
```

```python
#Lucro por Sub-categoria

profit_subcat = df.groupby("Sub-Category")["Profit"].sum().sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=profit_subcat.values, y=profit_subcat.index)
plt.title("Profit by Sub-Category")
plt.xlabel("Profit")
plt.ylabel("Sub-Category")
plt.show()
```

```python
# Selecionar apenas colunas numÃ©ricas para o cÃ¡lculo da correlaÃ§Ã£o
df_numeric = df.select_dtypes(include=['number'])

# Calcular a matriz de correlaÃ§Ã£o
correlation_matrix = df_numeric.corr()

# Criar um heatmap da matriz de correlaÃ§Ã£o
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Matriz de CorrelaÃ§Ã£o entre Medidas NumÃ©ricas')
plt.show()
```

```python
df["Order Month"] = df["Order Date"].dt.to_period("M")
```

```python

sales_month = df.groupby("Order Month")["Sales"].sum()
profit_month = df.groupby("Order Month")["Profit"].sum()

plt.figure(figsize=(12,5))

plt.plot(sales_month.index.astype(str), sales_month.values, label="Sales")
plt.plot(profit_month.index.astype(str), profit_month.values, label="Profit")

plt.title("Sales vs Profit Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Value")
plt.axhline(y=0, linestyle="--", alpha=0.7)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
